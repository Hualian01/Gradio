# Modificacion de Imagen usando interfaz de Gradio

```python
import numpy as np

def sepia(image):
    sepia_filter= np.array(
    [[0.393,0.769,0.189],
    [0.349,0.686,0.168],
     [0.272,0.534,0.131]
    ])
    
    sepia_img= image.dot(sepia_filter.T)
    sepia_img/=sepia_img.max()
    
    return sepia_img 
```

```python
import gradio as gr

gr.Interface(
    fn = sepia,    # funcion 
    inputs = gr.Image(), # tipo de entrada
    outputs = "image"  # tipo de salida, son clases definidas de gradio
).launch()
```

# Parametros de la Interface
 * **title**: agrega el titulo enla parte superior de la interface
 * **descrption**: Agrega una descripcion debajo del titulo, se puede utiliza Markdown y HTML
 * **examples**: permite entregar ejemplos de entrada
 * **inputs**: Puede ser una lista de tipos de entrada, los cuales son mapeados a la funci'on de prediccion
 * **outputs**: Puede ser una lista de tipos de salida, los cuales son mapeados de la salidad de la funcion de prediccion
 
 
 Gradio permite definir diferentes tipos de elementos para agregar en la interfaz web, en este caso se utiliza un radiobutton para permitir al usuario seleccionar entre una gama de opciones.

```python
def calculator(num1, operation, num2):
    if operation =='add':
        return num1+num2
    elif operation == 'substract':
        return num1-num2
    elif operation == 'multiply':
        return num1*num2
    elif operation == 'divide':
        return num1/num2
    

demo = gr.Interface(
    fn = calculator,
    inputs =[
        gr.Number(value =4),
        gr.Radio( ['add', 'substract', 'multiply', 'divide']  ),
        "number"
    ],
    outputs = "number",
    examples = [
        [5, "add", 3],
        [4,"divide",2],
        [-4, "multiply", 2.5],
        [0,"substract", 1.2]
        ],
    title ="Test Calculator",
    description = "Esto es una **calculadora de juguete**, creada con Gradio")

demo.launch()
```

# Block Para la Interfaz Grafica

Permite personalizar mas el diseno de la interfaz.
Se necesitan los mismos 3 elementos basicos.
Consideraciones:  
1. Blocks permite generar aplicaciones web que combinan markdown, HTML, botones y componentes interactivos simplemente instanciando objetos en Python dentro del contexto de with .Gradio. Blocks I ) . El orden en el que se instancian los componentes importan, ya que cada elemento es renderizado en la aplicación en el orden en el que fueron creados. 
2. Puedes definir funcones regulares de Python en cualquier parte de tu código y ejecutarlas usando Blocks. En el ejemplo, se define una función simple que retorna un mensaje de bienvenida, pero en la práctica es posible usar grandes modelos de ML. 
3. Es posible asignar eventos a los componentes de Blocks. Esto ejecutará tu función cuando el componente haya sido clickeado/cambiado/etc. Cuando se asigna un evento, se pasan tres parámetros: 
    - fn : la función a ser ejectuada 
    - inputs : lista de componentes de entrada
    - outputs : lista de componentes de salida 
 4. Blocks infiere de forma automática si un componente es interactivo (acepta input del usuario) o no, basado en los triggers definidos. En nuestro ejemplo, la primera entrada de texto no es interactiva, dado que el valor de salida no es usado como input. En algunos casos es deseable configurar esto, lo que es posible definiendo el booleano interactive, un parámetro que todo componente acepta. 
5. Es posible programar y desplegar Blocks en cualquier parte: jupyter notebooks, colab notebooks, o cualquier IDE de Python regular ya que Gradio usa el interpretrador estándar de Python. Es posible también compartir Blocks con otras personas seteando un simple parámetro: launch(share True) 



```python
def saludo(nombre):
    return f'bienvenido a Gradio Blocks, {nombre}!'


with gr.Blocks() as demo:
    gr.Markdown(
    """
    # Hola Mundo!
    Escribe algo debajo para el output.
    """
    )
    
    (placeholder="Cual es tu nombre?"inp = gr.Textbox)
    out = gr.Textbox("")
    
    inp.change(fn = saludo,  # con el metodo change, se hace que la 
               inputs=inp,   # que la funcion detecte un cambio y aplique
               outputs=out)  # la funcion de manera directa sin dar a un 
                             # boton. Es como un listener
    demo.launch()
```

# Uso de Tabs o Pestañas 

**¿Qué son las pestañas en Gradio?**

Las pestañas en Gradio son una forma de organizar diferentes interfaces de usuario dentro de una sola aplicación web. Permiten dividir una aplicación en secciones lógicas, lo que facilita la navegación y el uso.

**¿Cómo usar pestañas en Gradio?**

Para usar pestañas en Gradio, necesitas usar el componente Tabs. Este componente permite crear una fila de pestañas, cada una de las cuales representa una interfaz de usuario diferente.

```python
import numpy as no
import gradio as gr

demo = gr.Blocks()

def flip_text(x):
    return x[::-1]

def flip_image(image):
    return np.fliplr(image)


with demo:
    gr.Markdown("Flip *text* or *image* files using Python")
    with gr.Tabs():
        with gr.TabItem("Flip Text"):
            text_input =gr.Textbox()
            text_output=gr.Textbox()
            text_button =gr.Button("Flip")
        with gr.TabItem("Flip Image"):
            with gr.Row():
                image_input = gr.Image()
                image_output = gr.Image()
            image_button =gr.Button("Flip")
            
    text_button.click(flip_text, inputs=text_input, outputs=text_output)
    image_button.click(flip_image, inputs=image_input, outputs=image_output)
    
demo.launch()
```

```python
import altair

import gradio as gr
from math import sqrt
import matplotlib.pyplot as plt
import numpy as np
import plotly.express as px
import pandas as pd


def outbreak(plot_type, r, month, countries, social_distancing):
    months = ["January", "February", "March", "April", "May"]
    m = months.index(month)
    start_day = 30 * m
    final_day = 30 * (m + 1)
    x = np.arange(start_day, final_day + 1)
    pop_count = {"USA": 350, "Canada": 40, "Mexico": 300, "UK": 120}
    if social_distancing:
        r = sqrt(r)
    df = pd.DataFrame({"day": x})
    for country in countries:
        df[country] = x ** (r) * (pop_count[country] + 1)

    if plot_type == "Matplotlib":
        fig = plt.figure()
        plt.plot(df["day"], df[countries].to_numpy())
        plt.title("Outbreak in " + month)
        plt.ylabel("Cases")
        plt.xlabel("Days since Day 0")
        plt.legend(countries)
        return fig
    elif plot_type == "Plotly":
        fig = px.line(df, x="day", y=countries)
        fig.update_layout(
            title="Outbreak in " + month,
            xaxis_title="Cases",
            yaxis_title="Days Since Day 0",
        )
        return fig
    elif plot_type == "Altair":
        df = df.melt(id_vars="day").rename(columns={"variable": "country"})
        fig = altair.Chart(df).mark_line().encode(x="day", y='value', color='country')
        return fig
    else:
        raise ValueError("A plot type must be selected")


inputs = [
    gr.Dropdown(["Matplotlib", "Plotly", "Altair"], label="Plot Type"),
    gr.Slider(1, 4, 3.2, label="R"),
    gr.Dropdown(["January", "February", "March", "April", "May"], label="Month"),
    gr.CheckboxGroup(
        ["USA", "Canada", "Mexico", "UK"], label="Countries", value=["USA", "Canada"]
    ),
    gr.Checkbox(label="Social Distancing?"),
]
outputs = gr.Plot()

demo = gr.Interface(
    fn=outbreak,
    inputs=inputs,
    outputs=outputs,
    examples=[
        ["Matplotlib", 2, "March", ["Mexico", "UK"], True],
        ["Altair", 2, "March", ["Mexico", "Canada"], True],
        ["Plotly", 3.6, "February", ["Canada", "Mexico", "UK"], False],
    ],
    cache_examples=True,
)

if __name__ == "__main__":
    demo.launch()




	
```

# Problema 1

Realice una aplicacion con Gradio que reciba 3 elementos numericos como entrada y produzca 4 elementos de salida. 

- El primer elemento debe ser la suma de los 3 numeros
- Segundo elemento, la resta de todos los numeros
- El tercer elementos, La división del mayor de los numeros entre el menor de los numeros.

```python
import gradio as gr

def operaciones(numero1, numero2, numero3):  # función que realiza las 3 operaciones 
    suma_num = numero1 + numero2 + numero3
    resta_num = numero1 - numero2 - numero3
    if numero2 != 0 and numero3 != 0:   # verifica que no hay división por cero
        division_num = max(numero1, numero2, numero3) / min(numero1, numero2, numero3)
    else:
        division_num = "Error: División por cero"
    return f"La suma es: {suma_num}", f"La resta es: {resta_num}", f"La división es: {division_num}"  #Devuelve los resultados junto a un mensaje 

#creacion de la interfaz
interfaz = gr.Interface(
    fn=operaciones,
    inputs=["number", "number", "number"],
    outputs=["text", "text", "text"],
    title="Calculadora de Operaciones",
    description="Ingresa tres números y obtendra los resultados de suma, resta y división (del mayor de los numeros entre el menor.).",
)

interfaz.launch()
```

# Problema 2 

Defina una aplcacion con Gradio que reciba 2 listas de entre 10 a 15 numeros.
Utilizando Matplotlib, grafice las listas en un formato X-vs-Y. Muestre el resultado en el panel correspondiente.

```python
import gradio as gr
import matplotlib.pyplot as plt
import numpy as np

def plot_lists(lista1, lista2):
    # Convertir las listas de texto a listas de números
    try:
        list1 = [float(x.strip()) for x in lista1.split(",")]
        list2 = [float(y.strip()) for y in lista2.split(",")]
    except ValueError:
        return "Ingrese números válidos separados por comas."

    # Ajustar las listas para que tengan la misma longitud (rellenar con None si es necesario)
    max_length = max(len(list1), len(list2))
    list1 += [None] * (max_length - len(list1))
    list2 += [None] * (max_length - len(list2))

    # Crear un gráfico de dispersión X-vs-Y con líneas conectando los puntos 
    plt.figure(figsize=(8, 6))
    plt.plot(list1, list2, marker='o', linestyle='-', color='r', label='Datos')
    plt.xlabel('Lista 1 (X)')
    plt.ylabel('Lista 2 (Y)')
    plt.title('Gráfico de Dispersión X-vs-Y ')
    plt.grid(True)
    plt.legend()
    plt.tight_layout()

    # Guardar el gráfico en un archivo 
    plt.savefig('scatter_plot_connected_red.png')

    # Mostrar el gráfico 
    plt.show()

    return 'scatter_plot_connected_red.png'

# Crear la interfaz
interfaz = gr.Interface(
    fn=plot_lists,
    inputs=["text", "text"],
    outputs="image",
    title="Gráfico de Dispersión X-vs-Y",
    description="Ingrese dos  lista que tenga de  10 a 15 numeros separados por comas (Si es decimal ejemplo, 2.4) y observe el gráfico resultante.",
    allow_flagging=False
)

if __name__ == "__main__":
    interfaz.launch()

```

# Problema 3
Diseñe una interface en Gradio que reciba 2 imagenes y produzca 2 salidas de tipo imagen. La salida debe realizar algun tipo de modificación en cada imagen, de igual manera en la salida las imágenes deben intercambiar de posición, respecto a su ingreso inicial. Puede usar Numpy para las modificaciones o cualquier otra librería de modificación de imágenes.

```python
import gradio as gr
import numpy as np
from PIL import Image, ImageOps

def modify_images(imagen1, imagen2):
    # Convertir los datos de imagen de Gradio a una imagen PIL
    img1 = Image.fromarray(imagen1.astype('uint8'))
    img2 = Image.fromarray(imagen2.astype('uint8'))

    #  modificaciones 
    # 1. Convertir la primera imagen a blanco y negro
    modificacion_img1 = ImageOps.grayscale(img1)
    # 2. Agregar un borde a la segunda imagen
    modificacion_img2 = ImageOps.expand(img2, border=30, fill='black')

    # Intercambiar las imágenes modificadas
    output_img1 = np.array(modificacion_img2)
    output_img2 = np.array(modificacion_img1)

    return output_img1, output_img2

# Crear la interfaz de Gradio
interfaz = gr.Interface(
    fn=modify_images,
    inputs=["image", "image"],
    outputs=["image", "image"],
    title="Modificación y Cambio de Imágenes",
    description="Este programa toma dos imágenes como entrada y aplica modificaciones: la primera se convierte a blanco y negro, y la segunda se rodea de un borde negro.",
    allow_flagging=False
)

if __name__ == "__main__":
    interfaz.launch()
```

```python
import gradio as gr
from PIL import Image, ImageDraw, ImageFont
import tempfile

def texto_a_imagen(texto):
    # Creacón de la imagen que tenga un fondo blanco
    imagen = Image.new('RGB', (300, 100), color='white')
    draw = ImageDraw.Draw(imagen)
    
    # Tamaño de la fuente de la letra
    font_size = 24  # Tamaño de la fuente en puntos
    
    # Usar una fuente preinstalada y para el texto colo rojo
    fuente = ImageFont.truetype("arial.ttf", font_size)
    color_texto = 'red'
    
    # Dibujar el texto en la imagen, ajustando la posición
    draw.text((10, 30), texto, font=fuente, fill=color_texto)
    
    # Guardar la imagen en un archivo temporal
    temp_file = tempfile.NamedTemporaryFile(suffix='.png', delete=False)
    imagen.save(temp_file.name, format='PNG')
    
    return temp_file.name

# Crear la interfaz de Gradio
gr.Interface(fn=texto_a_imagen,
              inputs="text",
              outputs="image",
              title=" Aplicación#1 Trasladar un texto a una imagen.",
              description="Ingresa un texto corto para que se muestre en una imagen como salida.",
              theme="light").launch()



```

```python

import gradio as gr

def binario_a_decimal(numero_binario):   #Convierte un número binario a decimal
    conversion = int(numero_binario, 2)
    return conversion

def decimal_a_binario(numero_decimal):     #Convierte de dcimal a binario
    conversion = bin(int(numero_decimal))[2:]
    return conversion

def convertir(tipo_conversion, numero):
    if tipo_conversion == "bin":    # Ingresa bin se realizara la conversion de binario a decimal
        if all(bit in '01' for bit in numero):
            return binario_a_decimal(numero)
        else:
            return "No es un número binario válido."
    elif tipo_conversion == "dec":  # si ingresa dec se realizará la conversion decimal a binario
        return decimal_a_binario(numero)

interfaz = gr.Interface(           #Creacion de la interfaz 
    fn=convertir,
    inputs=["text", "text"],
    outputs="text",
    title="Aplicación #2  Convertir de Binario a Decimal / Decimal a Binario",
    description="Escribe 'bin'o  'dec' segun la operación que desea realizar, seguido ingrese el numero a  convertir.",
)

interfaz.launch()


#Utilice la primera interfaz (gradio.Interface(fn, inputs, outputs, ···)
```
